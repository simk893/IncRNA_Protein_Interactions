```python
import os
import pandas as pd
from IPython.display import display
data_directory = "C:\\Users\\simeo\\OneDrive\\Documents\\UNC\\Spring 2025\\STOR 765\\Inc_RNA_Protein_Predictions_Sample_Data"

# Dictionary of IncRNA-Protein Interaction Predictions Dataframes
dataframes_dict = {}
for filename in os.listdir(data_directory):
    file_path = os.path.join(data_directory, filename)
    if os.path.isfile(file_path) and filename.endswith(".tsv"):
        key = filename.replace("_predictions.tsv", "")
        df = pd.read_csv(file_path, sep='\t')
        dataframes_dict[key] = df
display(dataframes_dict)
```
